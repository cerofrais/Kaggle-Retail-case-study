```python
data=pd.read_csv('Train_Kaggle.csv')
data[data['Sales(In ThousandDollars)'].isnull()]
```

```python
for each in data[data['Sales(In ThousandDollars)'].isnull()].index.values.tolist():
    data.loc[each,'Sales(In ThousandDollars)']=0.5*(data.loc[each-3,'Sales(In ThousandDollars)']+data.loc[each+3,'Sales(In ThousandDollars)'])
```
